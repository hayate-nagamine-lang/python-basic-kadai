```python
class Human:

    def __init__(self,name,age):
        self.name = name
        self.age  = age

    def cheak_adult(self):
        if self.age >= 20 :
          print("大人です")

        else :
          print("大人ではありません")

human_list = []

human_list.append  (Human("侍太郎",19))
human_list.append  (Human("侍花子",21))
human_list.append  (Human("侍次郎",33))

for human in human_list:
    human.cheak_adult()
```
